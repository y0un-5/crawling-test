```python
# -*- encoding: utf-8 -*-
import requests
from bs4 import BeautifulSoup
import urllib.request
import csv
import os
import re
import calendar
from time import sleep
```

```python
url = "https://www.melon.com/chart/index.htm"
response = requests.get(url, headers = {"User-Agent":"Mozilla/5.0"})
response.headers['Content-Type']
text = response.text
html = BeautifulSoup(text, 'html.parser')
tr_list = html.find("tbody").find_all("tr")
len(tr_list)
```

```python
songs = []
names = []
albums = []
ranks = []
imgs = []
rank = 0
for tr in tr_list:
    rank += 1
    song = tr.find("div", {"class":"ellipsis rank01"}).find("a").text.strip()
    name = tr.find("div", {"class":"ellipsis rank02"}).find("a").text.strip()
    album = tr.find("div", {"class":"ellipsis rank03"}).find("a").text.strip()
    img = tr.find("a", {"class":"image_typeAll"}).find("img").get("src").replace('/120/', '/original/').replace('/80/', '/100/')
    
    songs.append(song)
    names.append(name)
    albums.append(album)
    ranks.append(rank)
    imgs.append(img)
    
    
total_ranks = []
for x, y, z, a in zip(ranks, songs, names, albums):
    total_rank = [x, y, z, a]
    total_ranks.append(total_rank)
    print(x, y, z, a)
```

```python
cnt = 0
for img in imgs:
    try:
        imgname = str(ranks[cnt]) + "_" + str(songs[cnt])
        imgname = re.sub('[-=.#/?:$\\n\\}]', '', imgname) + ".jpg"
        cnt += 1
        urllib.request.urlretrieve(img, f"img\\"+ imgname)
    except:
        imgname = str(ranks[cnt+1]) + "_" + str(songs[cnt+1])
        imgname = re.sub('[-=.#/[?:$\\n\\}]', '', imgname) + ".jpg"
        cnt += 1
        urllib.request.urlretrieve(img, f"img\\"+ imgname)
imgs
```
